```python
# importing the libraries
#from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
#Sending a HTTP request to a URL
url = "https://ceoharyana.gov.in/WebCMS/Start/1440"
# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).content
```

```python
df_list = pd.read_html(html_content)
df = df_list[-1]
print(df)
df.to_csv('assembly.csv')
```

```python
f=pd.read_csv("assembly.csv", skiprows=[i for i in range(1,79)])
keep_col = ['Division','District','No. & Name of Assembly Constituency']
assembly_constituencies = f[keep_col]
print(assembly_constituencies)
```

```python
#new_f.to_csv("assembly_constituencies.csv", index=False)
```

```python
assembly_constituencies["Division"].fillna("FARIDABAD", inplace = True)
assembly_constituencies
```

```python
ac= assembly_constituencies.replace({'Division': {'6 FARIDABAD': 'FARIDABAD'},'District': {'3-FARIDABAD': 'FARIDABAD','1-MEWAT':'MEWAT','2-PALWAL':'PALWAL'}})
ac
```

```python
ac.loc[[1,2], 'District'] = 'MEWAT'
ac.loc[[4,5], 'District'] = 'PALWAL'
ac.loc[[7,8,9,10,11], 'District'] = 'FARIDABAD'
ac
```

```python
ac['No. & Name of Assembly Constituency'] = ac['No. & Name of Assembly Constituency'].str[3:]
ac
```

```python
ac["location"] = ac["No. & Name of Assembly Constituency"] +", " + ac["District"] +", " + ac["Division"]
ac
```
